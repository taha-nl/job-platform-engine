```python
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import re
import csv
from itertools import zip_longest
import pandas as pd
```

```python
browser=webdriver.Chrome()


URL=f"https://www.glassdoor.com/Job/morocco-software-engineer-jobs-SRCH_IL.0,7_IN162_KO8,25.htm"

browser.get(URL)
time.sleep(2)

comapnies = [company.text for company in browser.find_elements(By.XPATH,'//div/div[@class="EmployerProfile_profileContainer__d5rMb"]')]
jobs = [job.text for job in browser.find_elements(By.XPATH,'//div/a[@class="JobCard_seoLink__WdqHZ"]')]
locations = [location.text for location in browser.find_elements(By.XPATH,'//div/div[@class="JobCard_location__N_iYE"]')]
# links = [ link.get_attribute('href') for link in  browser.find_elements(By.XPATH,'//a[@class="JobCard_trackingLink__zUSOo"]')]
links = browser.find_elements(By.XPATH,'//div[@class="JobCard_jobCardContainer__l0svv"]')
jobs_description = []
Founded_year = []
company_size = []
company_type = []
comany_industry = []
company_sector = []
Revenue = []




for link in links:
        
        browser.execute_script("arguments[0].click();", link)
        time.sleep(1)
          
        try:
                browser.find_element(By.XPATH,('//button[@class="CloseButton"]')).click() #clicking to the X.
                print(' x out worked')
        except:
                pass

        time.sleep(1)

        # job_description = browser.find_element(By.XPATH,'//div[@class="JobDetails_jobDescription__6VeBn JobDetails_blurDescription__fRQYh"]')
        # jobs_description.append(job_description.text)

        # try:
        about_company_keys = [key.text for key in browser.find_elements(By.XPATH,'//div[@class="JobDetails_overviewItem__35s2T"]/span')]
        about_company_values = [ value.text for value in browser.find_elements(By.XPATH,'//div[@class="JobDetails_overviewItem__35s2T"]/div')]
        about_company = dict(zip(about_company_keys,about_company_values))

        if about_company:

        
                Founded_year.append(about_company["Founded"] if "Founded" in about_company.keys() else None)
                company_size.append(about_company["Size"] if "Size" in about_company.keys() else None)
                company_type.append(about_company["Type"] if "Type" in about_company.keys() else None)
                comany_industry.append(about_company["Industry"] if "Industry" in about_company.keys() else None)
                company_sector.append(about_company["Sector"] if "Sector" in about_company.keys() else None)
                Revenue.append(about_company["Revenue"] if "Revenue" in about_company.keys() else None)
        else :

        
                Founded_year.append(None)
                company_size.append(None)
                company_type.append(None)
                comany_industry.append(None)
                company_sector.append (None)
                Revenue.append(None)







                







```

```python
dataframe = pd.DataFrame({
                        "company":comapnies[:1] + comapnies[2:],"job":jobs,"location":locations,
                          "company_size":company_size[:-1],"Founded_year":Founded_year[:-1],"company_type":company_type[:-1],
                          "comany_industry":comany_industry[:-1],"company_sector":company_sector[:-1],"Revenue":Revenue[:-1]
                          })

dataframe.head()
```

```python
dataframe.to_csv("./jobs_software_engineering_morocco.csv",index=False)
```
